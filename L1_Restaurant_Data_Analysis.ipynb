```python
# Basic statistical measures (mean, median, standard deviation, etc.) for numerical columns
# Select Numerical Columns
numeric_columns = df.select_dtypes(include = ['int', 'float'])

# Calculate basic statistical measures using .describe()
summary_stats = numeric_columns.describe()
print(summary_stats)
```

```python
# Calculate the median for numerical columns
median = numeric_columns.median()
print(f"\nMedian for numerical columns:\n{median}")
```

```python
# Calculate standard deviation for numerical columns
std_dev = numeric_columns.std()
print(f"\nStandard deviation for numerical columns:\n{std_dev}")
```

### Distribution of Categorical Variables

```python
# Distribution of categorical variables like 'Country Code', 'City', and 'Cuisines'
# Count plot visualization code for country codes
# Set plot size
plt.figure(figsize = (15, 6))

# Create the figure object
sns.countplot(x = df['Country Code'])

# Set Labels
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('Country Codes', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif')
plt.ylabel('Number of Restaurants', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title('Distribution of Restaurants by Country Codes', fontname = 'serif', fontsize = 16)

# Display the Chart
plt.show()
```

```python
# Count plot visualization code for cities
# Set plot size
plt.figure(figsize = (15, 6))

# Create the figure object
# There are many city names present in the data, so I selected only the top 10 cities
sns.countplot(x = df['City'], order = df.City.value_counts().iloc[:10].index)

# Set Labels
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('Name of Cities', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif', rotation = 90)
plt.ylabel('Number of Restaurants', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title('Top 10 Cities with the Highest Number of Restaurants', fontname = 'serif', fontsize = 16)

# Display the Chart
plt.show()
```

```python
# Count plot visualization code for cuisines
# Set plot size
plt.figure(figsize = (15, 6))

# Create the figure object
# There are many cuisine names present in the data, so I selected only the top 10 cuisines
ax = sns.countplot(x = df['Cuisines'], order = df.Cuisines.value_counts().iloc[:10].index)

# Set Labels
plt.grid(True, color = "grey", linewidth = "0.4")
plt.xlabel('Name of Cuisines', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif', rotation = 90)

plt.ylabel('Number of Restaurants', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.title('Top 10 Cuisines with the Highest Number of Restaurants', fontname = 'serif', fontsize = 16)

# Display the Chart
plt.show()
```

## **Task 3: Geospatial Analysis**

### Visualize Locations of Restaurants

```python
# Locations of restaurants on a map using latitude and longitude information
# Import all necessary libraries
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

# Create Point geometry from latitude and longitude using Shapely
gdf = gpd.GeoDataFrame(
    df,
    geometry = gpd.points_from_xy(df.Longitude, df.Latitude)
)

# Create a base map of the world using GeoPandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a map that fits the screen and plots the restaurant locations
# The 'continent' column is used for coloring, and a legend is displayed
gdf.plot(ax = world.plot("continent", legend = True, figsize = (14, 12)), marker = 'o', color = 'red', markersize = 16)

# Show the map
plt.show()
```

### Distribution of Restaurants by City

```python
# Distribution of restaurants across different cities or countries
# Set plot size
plt.figure(figsize = (15, 6))

# Create the figure object
# There are many city names present in the data, so I selected only the top 10 cities
sns.countplot(x = df['City'], order = df.City.value_counts().iloc[:10].index)

# Set Labels
plt.grid(True, color = "grey", linewidth = "0.4")
plt.ylabel('Number of Restaurants', fontname = 'serif', fontsize = 14)
plt.yticks(fontname = 'serif')
plt.xlabel('Name of Cities', fontname = 'serif', fontsize = 14)
plt.xticks(fontname = 'serif', rotation = 90)
plt.title('Distribution of Restaurants Across Cities', fontname = 'serif', fontsize = 16)

# Display the Chart
plt.show()
```

### Correlation Between the Restaurant's Location and its Rating

```python
# Checking the correlation between the restaurant's location and its rating
# Set plot size
plt.figure(figsize = (15, 6))

# Calculate the correlation between latitude, longitude, and ratings
correlation_matrix = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()

# Create a heatmap to visualize the correlation
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm', fmt = ".2f")

# Set Title
plt.xticks(fontname = 'serif')
plt.yticks(fontname = 'serif')
plt.title("Correlation Between Restaurant's location and Rating", fontname = 'serif', fontsize = 16)

# Display the Chart
plt.show()
```

### What did I find from Level 1 (Task 3)?

* North America and Asia (mainly India) have the most number of restaurants, followed by Oceania and others.
* New Delhi has the most number of restaurants, followed by Gurgaon, Noida, and Faridabad.
* There is no correlation between latitude and rating. However, longitude and rating are negatively correlated.
